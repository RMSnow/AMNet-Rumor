## Split

```python
# train:val = 4:1
#     train = 5064(0) + 5069(1) = 10133
#     val = 1266(0) + 1267(1) = 2533

import pandas as pd

df_0 = pd.read_csv('server_0_nonrumor.txt', header=None)
df_1 = pd.read_csv('server_1_rumor.txt', header=None)
```

```python
# 随机划分的次数
splits = 3
```

```python
for i in range(splits):
    df_0_shuffled = df_0.sample(frac=1).reset_index(drop=True)
    df_1_shuffled = df_1.sample(frac=1).reset_index(drop=True)

    train_0 = df_0_shuffled[:5064]
    val_0 = df_0_shuffled[5064:]
    train_1 = df_1_shuffled[:5069]
    val_1 = df_1_shuffled[5069:]
    
    train = pd.concat([train_0, train_1], ignore_index=True)
    val = pd.concat([val_0, val_1], ignore_index=True)
    
    # 选择shuffle
    train = train.sample(frac=1).reset_index(drop=True)
    val = val.sample(frac=1).reset_index(drop=True)
    
    train.to_csv('train_' + str(i)+ '.txt', header=None, index=None)
    val.to_csv('val_' + str(i)+ '.txt', header=None, index=None)
```
