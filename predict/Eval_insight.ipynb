# 分析 LSTM-Steps 的影响

```python
import pandas as pd
import matplotlib.pyplot as plt

img_path = 'eval_img/'
fig_size = (20,6)

expts_to_handle = []
for i in range(1, 5):
    expts_to_handle.append('expt' + str(i))
    expts_to_handle.append('expt' + str(i) + '.2')
```

## 全局对比

```python
# 选择expt1，expt2，expt3.2，expt4.2
lstm1_df = pd.read_csv('eval-expt2.csv')
lstm2_df = pd.read_csv('eval-expt3.2.csv')
lstm3_df = pd.read_csv('eval-expt1.csv')
lstm4_df = pd.read_csv('eval-expt4.2.csv')

dfs = pd.DataFrame(
    {'lstm1_pred':lstm1_df['PredValue'], 
     'lstm2_pred':lstm2_df['PredValue'], 
    'lstm3_pred':lstm3_df['PredValue'], 
    'lstm4_pred':lstm4_df['PredValue']}
)
```

```python
dfs.hist(figsize=fig_size)
plt.savefig(img_path + 'expts-hist.png')
plt.show()
```

```python
dfs.plot(kind='density', figsize=fig_size)
plt.savefig(img_path + 'expts-density.png')
plt.show()
```

## 各expt的分布图

```python
for expt_to_handle in expts_to_handle:
    eval_file = 'eval-' + expt_to_handle + '.csv'
    eval_df = pd.read_csv(eval_file)
    
    eval_df.hist(figsize=fig_size)
    plt.savefig(img_path + expt_to_handle + '-hist.png')
    eval_df.plot(kind='density', figsize=fig_size)
    plt.savefig(img_path + expt_to_handle + '-density.png')
    # plt.show()
```
